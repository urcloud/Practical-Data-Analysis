```python
import pandas as pd
import numpy as np
```

```python
df_train=pd.read_csv('mnist_train.csv')
df_test=pd.read_csv('mnist_test.csv')
```

```python
# Rename columns
pxl = ["pxl{}".format(i) for i in range(len(df_train.columns)-1)]
df_train.columns = ["value"] + pxl
df_test.columns =  ["value"] + pxl
```

```python
A_train = pd.DataFrame(df_train, columns = pxl)
y_data_train = pd.DataFrame(df_train, columns = ["value"]).values
A_test = pd.DataFrame(df_test, columns = pxl)
y_data_test = pd.DataFrame(df_test, columns = ["value"]).values
```

```python
def digit_to_vec(value):
  vec = np.zeros(shape = 10)
  if value != 0:
    vec[value-1] = 1
  else:
    vec[-1] = 1

  return vec
```

```python
y_data_train_vec = [digit_to_vec(y_data_train[k]) for k in range(len(y_data_train))]
B_train = np.array(y_data_train_vec) # encoding
y_data_test_vec = [digit_to_vec(y_data_test[k]) for k in range(len(y_data_test))]
B_test = np.array(y_data_test_vec) # encoding
```

```python
from tqdm.notebook import tqdm
eta = 1e-8
iteration = 100000

n = A_train.shape[0] # number of sample
k = A_train.shape[1] # number of feature
d = B_train.shape[1] # number of class
W = np.zeros((k, d)) # initial weight matrix

def loss(W): # loss: MSE
    E = A_train @ W - B_train
    return np.mean((E**2))

def grad(W): # gradient of loss
    return (A_train.T @ (A_train @ W - B_train))/n

# Gradient descent
for i in tqdm(range(iteration)): 
    G = grad(W)
    W = W - eta * G

    if i % 100 == 0:
        print("loss=", loss(W))
```

```python
B_predicted = A_test @ W 
round_B_predicted = np.round(B_predicted) 
num_correct = (round_B_predicted == B_test).all(axis=1)
accuracy = num_correct.mean()

print(f"accuracy: {accuracy:.4%}")
```
